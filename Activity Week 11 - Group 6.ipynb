# <font color='#ff7f24'>Preprocessing</font>

```python
y = df['passed']
x = df.drop(['passed'],axis=1)
print(y.head())
print(x.columns)
```

```python
def preProcess(a):
    newdf = pd.DataFrame(index = a.index)
    
    for col,colData in a.iteritems():
        if colData.dtype == object:
            colData = colData.replace(['yes','no'],[1,0])
        if colData.dtype == object:
            colData = pd.get_dummies(colData,prefix = col)
        
        newdf = newdf.join(colData)
    
    return newdf

x = preProcess(x)
x.head()
    
```

```python
x.shape
```

# <font color='#672e8c'>Splitting the Data</font>

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=42,test_size=0.2)
```

# <font color='#36a34e'>Training Model</font>

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix
```

```python
ptime=[]
ttime=[]
f1 = []
import time
```

```python
def model(m):
    a = m
    start = time.time()
    a.fit(xtrain,ytrain)
    end = time.time()
    print("{0}Model trained in {1} seconds".format(m,(end - start)))
    ttime.append(end - start)
    
    start = time.time()
    ypred = a.predict(xtest)
    end   = time.time()
    print("{0}Model Predicted in {1} seconds".format(m,(end - start)))
    ptime.append(end - start)
    
    print("F1 score of model {0} is :  {1}".format(m,f1_score(ypred,ytest,average='micro')))
    f1.append(f1_score(ypred,ytest,average='micro'))
    print("\n\n\n")
```

```python
mod = [LogisticRegression(random_state=42),RandomForestClassifier(random_state=42),
       DecisionTreeClassifier(random_state=42),SVC(kernel='linear',random_state=42),
      SVC(kernel='poly',random_state=42),SVC(kernel='rbf',random_state=42)]

for i in mod:
    model(i)
```

```python
mod = ['LogisticRegression','RandomForestClassifier',
       'DecisionTreeClassifier','SVC Linear',
      'SVC Poly','SVC-RBF']

newdf = pd.DataFrame(list(zip(mod, ttime,ptime,f1)),
               columns =['Model Name', 'Training Time','Prediction Time','F1 Score'])

newdf.plot(x="Model Name", y=['Training Time','Prediction Time','F1 Score'], kind="bar")
```

```python
# When we consider Training time prediction time and f1 score Decision tree is the best model

# When we only consider F1 score Logistic Regression is the best model
```
